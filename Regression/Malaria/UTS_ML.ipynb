1. Missing value treatment : dilakukan karena untuk memastikan tidak ada nilai null dalam dataframe

Missing Value Treatment

```python
df.isna().sum()[df.isna().sum()!=0]
```

Code diatas digunakan untuk memeriksa apakah ada null value di dalam dataframe. Hasil dari code diatas membuktikan bahwa tidak ada null value sehingga proses missing value treatment bisa dilewatkan.

Outliers treatment

```python
total_row = (len(df.columns)+2) // 3
fig, axes = plt.subplots(nrows = total_row, ncols = 3, figsize=(15,15))

for i, col in enumerate(df.columns):
  sns.boxplot(data = df, x = col, ax = axes[i//3, i%3])
plt.tight_layout()
plt.show()
```

Code diatas berfungsi untuk menampilkan boxplot dari setiap kolom. Dari boxplot tersebut dapat menentukan outlier yang perlu ditangani.

```python
outliers = df.quantile(.97)

for i in range(len(df.columns)):
  col = df.columns[i]
  df = df[(df[col]<outliers[col])]
```

```python
#cek lagi ada outliers tidak
total_row = (len(df.columns)+2) // 3
fig, axes = plt.subplots(nrows = total_row, ncols = 3, figsize=(15,15))

for i, col in enumerate(df.columns):
  sns.boxplot(data = df, x = col, ax = axes[i//3, i%3])
plt.tight_layout()
plt.show()
```

Normalisasi Skala Variabel

```python
X = df.iloc[:, :-1].values
Y = df.iloc[:, -1].values
```

```python
from sklearn.preprocessing import StandardScaler
stan = StandardScaler()
X = stan.fit_transform(df)
```

Code diatas digunakan untuk menormalisasi skala dari dataframe

Feature Extraction menggunakan PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
X_pca = pca.fit_transform(X)
explained_var_ratio = pca.explained_variance_ratio_
print(explained_var_ratio)
```

```python
pca = PCA(n_components = 15)
X_pca = pca.fit_transform(X)

important_features = pca.components_
print(important_features)
```

Code diatas digunakan untuk melakukan feature extraction dari variabel yang penting. Disini menggunakan PCA dengan jumlah komponen 15)

**D. Split dataset menjadi training dan testing**

```python
X = X_pca
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

**E. Model Linear Regression**

```python
from sklearn.linear_model import LinearRegression
regres = LinearRegression()
regres.fit(X_train, Y_train)
print("Koefisien : ", regres.coef_)
print("Intercept : ", regres.intercept_)
```

Melakukan prediksi dengan menggunakan test-set

```python
#prediction
Y_predict = regres.predict(X_test)
```

```python
print(regres.score(X_train, Y_train))
```

```python
print(regres.score(X_test, Y_test))
```

Menampilkan hasil prediksi dengan nilai sebenarnya

```python
df1 = pd.DataFrame(Y_test, columns=['Actual Value'])
df2 = pd.DataFrame(Y_predict, columns=['Predicted Value'])
concat_df = pd.concat([df1, df2], axis=1)
print(concat_df)
```

**F. Evaluasi Model Linear Regression**

Menampilkan regression model

```python
print("Koefisien : ", end=" ")
print(regres.coef_)
print("Intercept : ", end=" ")
print(regres.intercept_)
```

```python
plt.scatter(Y_test, Y_predict, color='red')
plt.plot(Y_test, Y_test, color='blue')
plt.xlabel("Actual Value")
plt.ylabel("Prediction Value")
plt.show()
```

Evaluasi Hasil Prediksi

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
rmse = np.sqrt(mean_squared_error(Y_test, Y_predict))
print("RMSE", end=" ")
print(rmse)
mse = mean_squared_error(Y_test, Y_predict)
print("MSE", end=" ")
print(mse)
mae = mean_absolute_error(Y_test, Y_predict)
print("MAE", end=" ")
print(mae)
r2 = r2_score(Y_test, Y_predict)
print("R2", end=" ")
print(r2)
```

Penjelasan mengenai hasil evaluation matrics

1. RMSE adalah evaluation matric yang mengukur akar dari rata-rata kuadrat selisih antara nilai aktual dan nilai prediksi
2. MSE adalah evaluation matric yang mengukur rata-rata kuadrat selisih antara nilai aktual dan nilai prediksi
3. MAE adalah evaluation matric yang mengukur rata-rata selisih absolut antara nilai aktual dan nilai prediksi pada data testing
4. R2 adalah evaluation matric yang mengukur kecocokan antara model linear regression dengan data training


Kesimpulan : Berdasarkan hasil evaluation matrics yang telah didapatkan bisa diambil kesimpulan bahwa model linear regression tersebut cukup bagus. Hal ini dikarenakan nilai RMSE dan MAE yang rendah yang mengindikasikan model dapat memprediksi nilai dengan akurat. Model juga menghasilkan nilai MSE yang rendah yang mengindikasikan model bida meng-generalisasi dengan baik. Selain itu, model juga menghasilkan nilai R2 yang tinggi yang mengindikasikan model bagus dalam merepresentasikan variasi data.
