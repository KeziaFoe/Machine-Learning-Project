Preprocessing

```python
#Data Cleaning
```

```python
#missing value
df.isna().sum()[df.isna().sum()!=0]
```

```python
#modus nya harus sesuai label penyakitnya
df['caling'].fillna(df[df['Disease'] == 4]['caling'].mode().values[0], inplace=True)
df['itching'].fillna(df[df['Disease'] == 1]['itching'].mode().values[0], inplace=True)

df.isna().sum()[df.isna().sum()!=0]
```

```python
#outliers
```

```python
#Data Construction
```

```python
#Labeling
#Di kolom Age, type datanya object
print(df['Age'].dtype)
#Ada '?' di beberapa data di kolom Age
#Hapus data yang tidak sesuai
df = df.drop(df.index[df['Age']=='?'])
df['Age'] = df['Age'].astype(int)
print(df['Age'].dtype)
```

```python
X = df.iloc[ : , :-1]
Y = df.iloc[ : , -1]

#Standarization
from sklearn.preprocessing import StandardScaler
stan = StandardScaler()
X = stan.fit_transform(X)
```

```python
#Train test split


from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#Oversampling
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=0)
x_sm, y_sm = sm.fit_resample(X_train, Y_train)
```

Modelling

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier

k=4
neigh = KNeighborsClassifier(n_neighbors=k).fit(x_sm, y_sm)

y_knn = neigh.predict(X_test)

from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_sm, neigh.predict(x_sm)))
print("Test set Accuracy: ", metrics.accuracy_score(Y_test, y_knn))
```

```python
Ks = 10
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(3,Ks):

    #Train Model and Predict
    neigh = KNeighborsClassifier(n_neighbors = n).fit(x_sm,y_sm)
    y_knn=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(Y_test, y_knn)


    std_acc[n-1]=np.std(y_knn==Y_test)/np.sqrt(y_knn.shape[0])

print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1)
```

```python
#Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(criterion='entropy', max_depth=5)
tree.fit(x_sm, y_sm)

predTree = tree.predict(X_test)
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(Y_test, predTree))
```

```python
#Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_sm, y_sm)
y_rf = rf.predict(X_test)
print("RandomForest's Accuracy: ", metrics.accuracy_score(Y_test, y_rf))
```

```python
#SVM
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel='linear', gamma=0.7, C=10)
clf.fit(x_sm, y_sm)
y_svm = clf.predict(X_test)
print("SVM's Accuracy: ", metrics.accuracy_score(Y_test, y_svm))
```

```python
# kalau pakai hipertuning
params = {'C':[0.5,1,2], 'kernel':['rbf','linear']}

svm = SVC()
from sklearn.metrics import fbeta_score, make_scorer
ftwo_scorer = make_scorer(fbeta_score, beta = 2)

from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator = svm, param_grid = params, n_jobs = 3, cv = 2)

from sklearn.model_selection import cross_val_score
print(cross_val_score(grid, x_sm, y_sm, cv=2))
```
